```python
import pandas as pd
import numpy as np

data = pd.read_csv(r'worldwide_state_level_data_covid_4_17_2022.csv').dropna(axis=0, how='all')

pd.options.display.max_rows = 10

data
```

```python
data1 = data.replace(np.nan, 0)
data1
```

```python
data_filter1 = data1.filter(items=('date', 'people_vaccinated', 'people_fully_vaccinated', 'deaths', 'administrative_area_level_1', 'administrative_area_level_2', 'iso_alpha_3', 'iso_alpha_2'), axis='columns')
data_filter1
```

```python
data_filter1_list = data_filter1['date'].tolist()
data_filter1_list
```

```python
data_filter2 = data_filter1['administrative_area_level_1']=='United States'
data_filter2.head()
```

```python
data_filter2A = data_filter1[data_filter2]
data_filter2A
```

```python
unique_state_values = data_filter2A['administrative_area_level_2'].unique()
unique_state_values
```

```python
data_filter3 = data_filter2A.filter(items=('date', 'people_fully_vaccinated', 'deaths', 'administrative_area_level_2'))
data_filter3
```

```python
data_filter4 = data_filter3.filter(items=('date', 'deaths', 'administrative_area_level_2'))
data_filter4
```

```python
data_filter5 = data_filter3.filter(items=('date', 'people_fully_vaccinated', 'administrative_area_level_2'))
data_filter5
```

```python
data_filter4.sort_values('date')
data_filter5.sort_values('date')
```

```python
data_filter4_pivot = data_filter4.pivot(index='administrative_area_level_2', columns='date', values='deaths')
data_filter4_pivot
```

```python
data_filter5_pivot = data_filter5.pivot(index='administrative_area_level_2', columns='date', values='people_fully_vaccinated')
data_filter5_pivot
```
